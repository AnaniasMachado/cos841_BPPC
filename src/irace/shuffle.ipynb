```python
import random
```

```python
def shuffle_instances(instances_file,
                      out_instances="stratified_sample_shuffled.txt",
                      seed=None):

    # Optional reproducibility
    if seed is not None:
        random.seed(seed)

    # ---- Read instances ----
    with open(instances_file, "r") as f:
        instances = [line.strip() for line in f if line.strip()]

    # ---- Shuffle ----
    random.shuffle(instances)

    # ---- Write shuffled instances ----
    with open(out_instances, "w") as f:
        for inst in instances:
            f.write(inst + "\n")

    print("Shuffling completed.")
    print(f"Instances to {out_instances}")
```

```python
shuffle_instances("stratified_sample.txt")
```
